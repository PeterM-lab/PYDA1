```python
data_ = pd.read_csv('power.csv')
```

```python
data_.head()
```

```python
category = [4, 12, 21]
countries = ['Latvia', 'Lithuania', 'Estonia']
```

```python
data_[ (data_.country.isin(countries)) &
      (data_.category.isin(category)) &
      (data_.year >= 2005) &
      (data_.year <= 2010) &
     (data_.quantity>0)]['quantity'].sum()
```
